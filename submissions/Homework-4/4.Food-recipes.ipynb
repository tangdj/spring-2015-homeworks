```python
import time
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
import sklearn.datasets as datasets
import sklearn.metrics as metrics
import scipy.sparse.linalg as linalg
import scipy.cluster.hierarchy as hr
import sklearn.cluster as cluster
from sklearn.decomposition import TruncatedSVD
from scipy.spatial.distance import pdist, squareform
from sklearn.preprocessing import StandardScaler

#import matplotlib as mpl
import seaborn as sns
%matplotlib inline
```

## Data parsing and representation

```python
allr = open("allr_recipes.txt", "rb")
epic=open("allr_recipes.txt","rb")
menu=open("menu_recipes.txt","rb")
maplist=open("map.txt","rb")
lists=[]
recipe=[]
maps=[]
country=[]
for line in allr:
    lists.append(line.split()[0])
    recipe.append(" ".join(line.split()[1:-1]))
for line in epic:
    lists.append(line.split()[0])
    recipe.append(" ".join(line.split()[1:-1]))
for line in menu:
    lists.append(line.split()[0])
    recipe.append(" ".join(line.split()[1:-1]))
for line in maplist:
    if len(line.split())==2:
        country.append(line.split()[0])
        maps.append(line.split()[1])
recipes=pd.DataFrame({'Countries':lists,'Recipes':recipe})
locations=pd.DataFrame({'Countries':country,'Regions':maps})
#recipes=pd.read_csv(f)
recipes=pd.merge(recipes,locations,on='Countries',how='left')
print recipes
```

## Clustering

```python
from sklearn.feature_extraction.text import TfidfVectorizer
countries=[]
countries=recipes["Countries"]
ingrediants=[]
lists=""
countries=list(set(countries))
for items in  recipes["Recipes"]:
    ingrediants.append(items)
vectorizer = TfidfVectorizer(stop_words='english', min_df=0.05, max_df=0.9)
vectors=vectorizer.fit_transform(ingrediants)
#Z = hr.linkage(vectors.todense(), method='complete', metric='euclidean')
print vectors.shape
#print type(vectors),vectors.shape
```

```python
U,s,V = linalg.svds(vectors,2,which='LM')
print U.shape, V.shape, s.shape
print s[::-1]
```

```python
Xk = U*sp.sparse.diags(s,0)
Z = hr.linkage(Xk[:2000], method='centroid', metric='euclidean')
fig = plt.figure(figsize=(10,10))
T = hr.dendrogram(Z,color_threshold=0.4, leaf_font_size=4)
fig.show()
```

Based on the graph shown above, we decide 11 clusters to do clustering

```python
kmeans = cluster.KMeans(init='k-means++', n_clusters=11, n_init=10)
kmeans.fit_predict(Xk)
labels = kmeans.labels_
fig, ax = plt.subplots(1,1,figsize=(10,10))
cmap = ['aliceblue','black','purple','orange','darkblue','pink','red','gray','green','hotpink','gold']
for i, label in enumerate(set(labels)):
    point_indices = np.where(labels == label)[0]
    point_indices = point_indices.tolist()
    plt.scatter(Xk[point_indices,0], Xk[point_indices,1], s=20, alpha=0.5, c=cmap[i], marker='D')
    plt.legend(prop={'size':14}, loc=2)
```

We find that when we decrease the number of singular values into 2. It has a very good performance on clustering and we want to find the two feature names.

```python
print("Top terms per cluster:")
asc_order_centroids = kmeans.cluster_centers_.argsort()#[:, ::-1]
order_centroids = asc_order_centroids[:,::-1]
terms = vectorizer.get_feature_names()
for i in range(11):
    print "Cluster %d:" % i
    for ind in order_centroids[i, :2]:
        print ' %s' % terms[ind]
    print
```
