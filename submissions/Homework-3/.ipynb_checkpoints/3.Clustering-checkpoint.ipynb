## Step 1. Simple clustering

The first thing that we will try is to cluster the geolocation data of the calls. The geolocation of a record consists of a pair `(latitude, longitude)`, which represents the position of the record on the map. Since the dataset that we have spans only a small geographic location, i.e. Boston, we will use the Euclidean distance when measuring how close two points are. Techinically, we should pick a [projection](https://github.com/mbostock/d3/wiki/Geo-Projections) in order to transform the data into 2d, and measure  the Euclidean distance in the projected space.

Since we are dealing with numeric data, we might consider centering and normalizing them. In order to do that, check the [StandardScaler](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler) from `scikit-learn`.

For exampe, the result of a clustering with 10 clusters is the following:


![A clustering of the locations of the calls for k=10](clusters10.png)

Filter the dataset to get the latitude and longitude, transform it as required and run the k-means clustering algorithm. Argue about the number of clusters you used. **(pts 50)**

```python
from sklearn import preprocessing
import numpy as np
import sklearn.metrics as metrics
df=boston[['LATITUDE','LONGITUDE']]
scalerX=preprocessing.StandardScaler().fit(df)
df=scalerX.transform(df)
print df
```

Firstly, we want to have a general idea how this distribution looks like so that we want to find the perfect clustering manually and then set it as a ground truth to be served as future standard to justify whether our later clustering is OK or not. 

```python
from mpl_toolkits.basemap import Basemap
bs=boston[['LATITUDE','LONGITUDE']]
temp=[]
for row in bs.iterrows():
    index,data=row
    temp.append(data.tolist())
m=Basemap(projection='mill',llcrnrlat=42.4594,urcrnrlat=42.1654,llcrnrlon=-71.3246,urcrnrlon=-70.6716,resolution='c')
m.drawcoastlines()
m.drawcountries()
m.drawstates()
m.fillcontinents(color='#04BAE3',lake_color='#FFFFFF')
plt.title('Boston distributions!')
for i in range(len(temp)):
    lat,lon=temp[i][0],temp[i][1]
    x,y=m(lon,lat)
    m.plot(x,y,'r.')
plt.show()
```

However, it is really hard for us to do any judgement just from the map above, because all the points are very close from each other. Thus, what we do next is to through a specific distance value to judge how far between each point

```python
def sc_evaluate_clusters(X,max_clusters):
    s = np.zeros(max_clusters+1)
    s[0] = 0;
    s[1] = 0;
    for k in range(2,max_clusters+1):
        kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
        kmeans.fit_predict(X)
        s[k] = metrics.silhouette_score(X,kmeans.labels_,metric='euclidean')
    plt.plot(range(2,len(s)),s[2:])
    plt.xlabel('Number of clusters')
    plt.ylabel('Adjusted Rand Index')
    
sc_evaluate_clusters(df[30000:40000],25)
```

It is impossible to calculate all the euclidean distances for all data. I choose random set of data and calculate the rand index for several times. there are few results whose values exceed 0.5. Thus, from Silhouette Coefficient, it is kind of hard for us to  decide the number of clusters. Later, we use the inertia to find the number of clusters.

```python
error = np.zeros(40)
error[0] = 0;
for k in range(1,40):
    kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
    kmeans.fit_predict(df)
    error[k] = kmeans.inertia_

plt.plot(range(1,len(error)),error[1:])
plt.xlabel('Number of clusters')
plt.ylabel('Error')
```

Based on the graph above, we decide use 15 clusters

```python
fig, ax = plt.subplots(1,1,figsize=(10,10))
cmap = sns.color_palette("bright", n_colors=4)
cmap.extend(sns.hls_palette(n_colors=11, h=0.7, l=0.4, s=0.8))
kmeans = KMeans(init='k-means++', n_clusters=15, n_init=10)
kmeans.fit_predict(df)
labels = kmeans.labels_
for i, label in enumerate(set(labels)):
    point_indices = np.where(labels == label)[0]
    point_indices = point_indices.tolist()
    plt.scatter(df[point_indices,1], df[point_indices,0], s=25, alpha=0.2, c=cmap[i], marker='+')
```
